```python
import os
import shutil
```

```python
testDir = "C:\\Users\\austi\\Documents\\ML Work\\Kitty Cam\\test_images"
destinationDir = "C:\\Users\\austi\\Documents\\ML Work\\Kitty Cam\\destination"
```

```python
def copyImages(srcDir, destDir):
    cnt = 0
    for root, dirs, files in os.walk(srcDir):
        for name in files:
            print(os.path.join(root, name))
            if os.path.join(root, name)[-4:] == ".jpg":
                shutil.copy(os.path.join(root, name), os.path.join(destDir, f"{cnt}.jpg"))
                cnt += 1
```

```python
copyImages(testDir, destinationDir)
```
